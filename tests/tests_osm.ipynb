```python
def osm_query_string_by_id(interest_area_code,interest_tag="building",node=False,way=True,relation=True,print_querystring=False):

    node_part = way_part = relation_part = ''

    if node:
        node_part = f'node["{interest_tag}"](area:{interest_area_code});'
    if way:
        way_part = f'way["{interest_tag}"](area:{interest_area_code});'
    if relation:
        relation_part = f'relation["{interest_tag}"](area:{interest_area_code});'

    overpass_query = f"""
    (  
        {node_part}
        {way_part}
        {relation_part}
    );
    /*added by auto repair*/
    (._;>;);
    /*end of auto repair*/
    out;
    """

    if print_querystring:
        print(overpass_query)

    return overpass_query

query_pb2 = osm_query_string_by_id(pb_code,print_querystring=True)
```
